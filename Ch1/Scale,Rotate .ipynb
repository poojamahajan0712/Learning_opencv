```python
import numpy as np
import cv2
import os 
os.chdir("D:\Pooja\Online Courses Content\Lynda\opencv\Ch2")
import matplotlib.pyplot as plt
```

```python
img= cv2.imread("Frog.jpg",1)
```

```python
plt.axis("off")
plt.title("Original")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) #converting from cv2 format for matplotlib
img.shape
```

Resizing 

```python
img_half=cv2.resize(img,(0,0),fx=0.5,fy=0.5)  ##fx ,fy - scaling factors for x and y axis 
img_stretch=cv2.resize(img,(600,600))
img_stretch_near=cv2.resize(img,(600,600),interpolation=cv2.INTER_NEAREST) #nearest neighbour interpolation
```

```python
cv2.imshow("Half",img_half)
cv2.imshow("Stretch",img_stretch)
cv2.imshow("Stretch Near",img_stretch_near)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

Shapes of Resized images

```python
print('Size of halved image',img_half.shape)
print('Size of Stretched image',img_stretch.shape)
print('Size of Stretched image using NN interpolation',img_stretch_near.shape) #will be same as above 
```

Rotation 

```python
M=cv2.getRotationMatrix2D((0,0),10,1)  # center,angle,scale 10 degree rotation from origin
rotated= cv2.warpAffine(img,M,(img.shape[1],img.shape[0]))  #warpAffine transforms the source image using the specified matrix
```

```python
cv2.imshow('Rotated',rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
plt.axis("off")
plt.title("Rotated")
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)) #converting from cv2 format for matplotlib
img.shape
```
