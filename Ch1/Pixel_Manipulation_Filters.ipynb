```python
# second parameter -> 1 - read colored image , 0 - grayscale
color=cv2.imread('min.jpg',1)
plt.axis("off")
plt.title("Original Image")
plt.imshow(cv2.cvtColor(color, cv2.COLOR_BGR2RGB))
```

Convert to grayscale image 

```python
#explicitly convert to gray image 
gray= cv2.cvtColor(color,cv2.COLOR_RGB2GRAY)
plt.axis("off")
plt.title("Converted to Gray ")
plt.imshow(gray,cmap=plt.get_cmap('gray'))
```

```python
#saving gray image using imwrite 
cv2.imwrite("gray.jpg",gray)
```

#Adding another channel for transparency 

```python
#splitting into channels 
b=color[:,:,0]
g=color[:,:,1]
r=color[:,:,2]
#we can use cv2.split also for the above task 
```

```python
rgba= cv2.merge((b,g,r,r)) #merging back , using r channel as transparency channel
```

```python
plt.axis("off")
plt.title("Added alpha channel ")
plt.imshow(rgba)
```

```python
cv2.imwrite("rgba.png",rgba)  # save in png format as jpg doesnt support image transparency 
```

# Blur , Dilation and Erosion

These operations help reduce noise or unwanted variances of an image or threshold

Gaussian Blur - smooths an image by averaging pixel values with its neighbours , to mooth out noise in an image 

```python
bde_b=cv2.imread('bde.jpg')
```

```python
blur=cv2.GaussianBlur(bde_b,(5,55),0)
```

for dilation and erosion we need a kernel 

```python
kernel = np.ones((5,5),'uint8')
```

Dilation effect works to turn black pixels / background pixels into white pixels 

```python
dilate=cv2.dilate(bde_b,kernel,iterations=1) # iterations tells the no. of times this operation should be applied
```

Erosion looks to turn white pixels into black

```python
erode=cv2.erode(bde_b,kernel,iterations=1)
```

```python
##viewing all using cv2
cv2.imshow("Original",bde_b)
cv2.imshow("Blur",blur)
cv2.imshow("dilate",dilate)
cv2.imshow("erode",erode)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

Viewing inline using matplotlib

```python

#subplot(r,c) provide the no. of rows and columns
fig, ax = plt.subplots(1,4) 
fig.set_figheight(20)
fig.set_figwidth(20)

# use the created array to output your multiple images. In this case I have stacked 4 images vertically
ax[0].imshow(cv2.cvtColor(bde_b, cv2.COLOR_BGR2RGB))
ax[1].imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
ax[2].imshow(cv2.cvtColor(dilate, cv2.COLOR_BGR2RGB))
ax[3].imshow(cv2.cvtColor(erode, cv2.COLOR_BGR2RGB))

ax[0].title.set_text('Original')
ax[1].title.set_text('Gaussian Blur')
ax[2].title.set_text('Dilation')
ax[3].title.set_text('Erode')

for ax1 in ax:
    ax1.set_xticks([])
    ax1.set_yticks([])
```
