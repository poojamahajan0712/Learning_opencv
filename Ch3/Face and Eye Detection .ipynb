# Haar cascade method 

feature based machine learning  , uses pretrained images of labeled positives and negatives . Runs through thousands of classifiers in a cascaded manner - detects faces in the image and draw bounding boxes

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os 
os.chdir("D:/Pooja/Online Courses Content/Lynda/opencv/Ch 3")
```

```python
img = cv2.imread("img2.jpg", 1) 
img1=img.copy()
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.subplot(121),plt.imshow(img)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
```

detectMultiscale module of the classifier - This function will return a rectangle with coordinates(x,y,w,h) around the detected face. <br>
Here is a list of the most common parameters of the detectMultiScale function :<br>
scaleFactor : Parameter specifying how much the image size is reduced at each image scale.<br>
minNeighbors : Parameter specifying how many neighbors each candidate rectangle should have to retain it.<br>
minSize : Minimum possible object size. Objects smaller than that are ignored.<br>
maxSize : Maximum possible object size. Objects larger than that are ignored.

```python
path = "haarcascade_frontalface_default.xml"
face_cascade = cv2.CascadeClassifier(path)
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=5, minSize=(40,40))
print (len(faces))
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x,y), (x+w,y+h), (2,255,0), 2) 
    
cv2.imshow("Image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
plt.subplot(121),plt.imshow(img)
plt.title('Face Detection'), plt.xticks([]), plt.yticks([])
```

```python
                                     
eye_cascade = cv2.CascadeClassifier("haarcascade_eye.xml")

eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.03, minNeighbors=10, minSize=(10,10))
print (len(eyes))

for (x, y, w, h) in eyes:
    cv2.rectangle(img1, (x,y), (x+w,y+h), (255,0,0),1) 
    
    
cv2.imshow("Eye detection ",img1)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
